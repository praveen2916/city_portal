```python
def geonameid_scraping(df):    
    api_key='AIzaSyDr1wlpuQrUd4fz26B-hi9SJh4cAFEXP_8'
    user_name = 'TerryYin1777'
    for i in range(df.shape[0]):#match_uniques_nan_joined.shape[0]):

        city_name = df.loc[i,'NAME']
        geocoding_url='https://maps.googleapis.com/maps/api/geocode/json?address='+city_name+',France'+'&key='+api_key

        try:
            geocoding_request = requests.get(geocoding_url)
            geocoding_results = geocoding_request.json()
            lat = geocoding_results['results'][0]['geometry']['location']['lat']
            long = geocoding_results['results'][0]['geometry']['location']['lng']

            geonameid_url = 'http://api.geonames.org/findNearbyPlaceNameJSON?lat='+str(lat)+'&lng='+str(long)+'&username='+user_name
            geonameid_request = requests.get(geonameid_url)
            results_geoid = geonameid_request.json()
            matched_name = results_geoid['geonames'][0]['name']
            population = results_geoid['geonames'][0]['population']
            geonameid = results_geoid['geonames'][0]['geonameId']

            df.loc[i,['scraped_name','scraped_population','scraped_geonameid']] = [matched_name,population,geonameid]

        except:
            print('error happened when checking {}'.format(city_name))

        if i % 10 == 0:
            print(i, 'cities done!')
```

```python
common_unique_cities = list(set(Census_dups['NAME']).intersection(france_dups['name']))
```

```python
#census_unique_sub = pd.DataFrame(columns = ['COM','NAME','POP_15'])
count = 0
for c in common_unique_cities:
    #print(c)
    census_sub = Census_dups[Census_dups['NAME']==c]
    census_sub_indexes = census_sub.index
    census_sub_len = len(census_sub_indexes)
    #print(census_sub_indexes)
    
    france_sub = france_dups[france_dups['name']==c]
    france_sub_indexes = france_sub.index
    france_sub_len = len(france_sub_indexes)
    
    if census_sub_len == 0 or france_sub_len == 0:
        continue
    else:
        to_be_encoded = np.min([census_sub_len,france_sub_len])
        
    for i in range(to_be_encoded):
        #print(to_be_encoded)
        census_sub_index = census_sub_indexes[i]
        france_sub_index = france_sub_indexes[i]
        #print(census_sub_index)
        Census_dups.loc[census_sub_index,'NAME'] = Census_dups.loc[census_sub_index,'NAME']+'_'+str(i)
        france_dups.loc[france_sub_index,'name'] = france_dups.loc[france_sub_index,'name']+'_'+str(i) 
        
    count += 1
    if count % 50 == 0:
        print('{} done!'.format(count))
```
